```python
import requests #匯入函式庫
from bs4 import BeautifulSoup #大倉庫匯bs4入小倉庫BeautifulSoup

res = requests.get("https://www.ptt.cc/bbs/Food/index7003.html")
#從requests中有一個方法get，告他要取得哪一個網站的的資料，把回傳直放入res

soup = BeautifulSoup(res.text,'html.parser')
#解析網頁內容，res.text指回傳內容文字的部分，html.parser解析的模式

articles = soup.select('.title a')
#要去網站的開發人員模式看，要先定錨class要用.代表，空一格代表下一層，齊一性:把相同格式內容的東西爬取回來

for each_title in articles:#透過迴圈把程式碼的部分拿掉
    print(each_title.text)#指列印文字部分
```
