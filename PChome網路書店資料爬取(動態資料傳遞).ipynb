```python
#看得到但爬不到
import requests
import json
from jsonpath import jsonpath

x = 1
for offset in range(5):#5頁
    url = "https://ecapi.pchome.com.tw/cdn/ecshop/prodapi/v2/region/DJAD/salesrank/0/prod&offset=x&limit=20&fields=Id,Nick,Pic,Price,Discount,isSpec,Name,isCarrier,isSnapUp,isBigCart&_callback=jsonp_prodlist&1614346978578?_callback=jsonp_prodlist"
    url = url.replace('x', str(x))#前者(網址上的x)被後者替代，迴圈每跑一次x+20
    res = requests.get(url)#傳入最新替換好的url
    #這邊不做網頁解析beautifulsoup(因為response回來的資料不是用html撰寫而是json格式 所以用動態解析方式來解析json型態的資料)
    clean_needed = res.text#有部分內容不符合json檔格式
    clean_needed
    #取代掉不符合json的格式(用2次取代頭尾)
    clean = clean_needed.replace('try{jsonp_prodlist(','').replace(');}catch(e){if(window.console){console.log(e);}}','')
    Json_Data = json.loads(clean)#編碼成人類看得懂的json(用json函式庫)
    
    for times in range(19):
        #跳進json層級中提取資料(用jsonpath函式庫)，$.查詢欄位
        book_name = jsonpath(Json_Data[times], '$.Nick')
        book_price = jsonpath(Json_Data[times], '$.Price.P')#json中的json
        print(book_name + book_price)
        
    x = x+20
```
