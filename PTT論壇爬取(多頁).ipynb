```python
import requests #匯入函式庫
from bs4 import BeautifulSoup #大倉庫匯bs4入小倉庫BeautifulSoup

url = "https://www.ptt.cc/bbs/Food/index.html"#最新的頁面開始爬

for times in range(50):#爬50頁
    res = requests.get(url)
    soup = BeautifulSoup(res.text,'html.parser')#解析爬回來的頁面資料
    articles = soup.select('.title a')#選取爬回來的資料
    paging = soup.select('.btn-group.btn-group-paging a')#要去網站的開發人員模式看，要選取的上一層名稱，空格需補上點代表不同動作
    #包括上頁、下頁...
    last_url = "https://www.ptt.cc" + paging[1]['href']#前一頁的網址
    url = last_url
    for each_title in articles:
        print(each_title.text,"https://www.ptt.cc" + each_title['href'])
```
